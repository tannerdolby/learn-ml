```python
x_train = np.array([1.464, 2.39, 1.296, 2.838, 2.370, 2.362, 1.343, 1.966, 1.974, 2.080, .974, 1.334, 1.822, 2.853])
y_train = np.array([215, 635, 453.990, 1100, 673, 775, 559.90, 651, 565, 765.873, 452.5, 480, 730, 1206.50])
m = x_train.shape[0]

# Take a dataset and create a table in Markdown

def createMarkdownTable(headers, columns, numRows):
    """
    Create a table in Markdown

    Parameters:
        headers (list | ndarray (m,)) : List of table headers
        columns (list | ndarray(m,))  : Table column data
        numRows (number)              : Number of table rows e.g. length of each column
    Returns:
        table   (str)                 : Markdown table string
    """
    table = ""

    # Create headers row and separate column headers by "|"
    # then create the next line which separates the headers
    # from the table data
    headerRow = "| "
    headerRowSep = "|"
    for header in headers:
        headerRow += header + " | "
        headerRowSep += "-" * (len(header)+2) + "|"

    rows = ""
    spacer = (len(headerRow)-3) // len(headers)
    
    # Construct rows
    headerIdx = 0
    for i in range(numRows):
        for column in columns:
            headerIdx %= len(headers)
            rows += "| " + str(column[i]) + " " * (len(headers[headerIdx]) - len(str(column[i])) + 1)
            if (headerIdx is len(headers)-1):
                rows += "|\n"
            headerIdx += 1

    return f"{headerRow}\n{headerRowSep}\n{rows}"
    
        

headers = list(['Size (1000 sqft)', 'Price (1000s of dollars)', 'TEst Header'])
cols = list([x_train, y_train, x_train])
markdownTable = createMarkdownTable(headers, cols, len(x_train))

print(markdownTable)
print(f"Training Examples: {m}")
for i in range(m):
    print(f"{x_train[i]*1000} sqft house sold for ${y_train[i]*1000} - ({x_train[i]}, {y_train[i]})")
```

The linear regression model will take an array of input features `x` (house price in sqft) and output a prediction for `y` which is represented as y-hat. 

Our function to create a straight line through the dataset will take two parameters `w` and `b`: F<sub>w,b</sub> (x<sup>(i)</sup>) = w * x + b

```python
w = 300
b = 50
```

```python
def compute_model_output(x, w, b):
    """
    Compute house price prediction of a linear model
    Args:
        x (ndarray (m,)) : Data, m examples
        w,b (scalar)     : parameters
    Returns:
        f_wb (ndarray (m,)) : predictions
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
plt.scatter(x_train, y_train, marker='x', c='red')
plt.title("Training Set")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (in 1000s of sqft)")
plt.show()
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
print(tmp_f_wb)
```

Let's draw a straight blue line through the training set data using our model output to see if the parameters need any tuning. Then we can begin using the model for house price predictions based on an input house size in sqft

```python
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (in 1000s of sqft)")
plt.legend()
plt.show()
```

Since the model is only using a single variable (house size in sqft), we aren't taking into account the number of bedrooms, bathrooms, lot size, or area which can have a large impact on the overall price of the house.

That being said, our imaginary client has a house size of 1750 sqft and wants to know a good price to initially list it on the market. Let's use our model and see what we can find out.

```python
client_house_size = np.array([1.75])
prediction = compute_model_output(client_house_size, w, b)
print(f"A house size of {client_house_size[0]*1000} sqft can sell for approximately ${prediction[0]*1000} in the current market")
```
